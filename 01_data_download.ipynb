```python
###############################################################################
##
## Name: 01_data_download.ipynb
## Purpose: 
## Author: VDR
## Email: dellarocca.vale@gmail.com
## Date: 2021-12-05
## Version: v1
##
###############################################################################
##
## Notes: Same output of what provided in April, just rewritten in Python
##
###############################################################################



## 0. General set up ----------------------------------------------------------



# Libraries
import os
import pandas as pd
import numpy as np
from pathlib import Path
import re
import requests
from urllib.request import urlopen, urlretrieve

os.getcwd()

## Create subfolder for saving the files if it doesn't exist already
Path("raw_data").mkdir(parents=True, exist_ok=True)


# 0.1 Choice of the data sources =======================================================================



url_name_list = ["https://www.ons.gov.uk/economy/economicoutputandproductivity/output/datasets/ukspendingoncreditanddebitcards/",
                  "https://www.ons.gov.uk/economy/economicoutputandproductivity/output/datasets/indexofservices/current/",
                  "https://www.gov.uk/government/uploads/system/uploads/attachment_data/file/972000/CJRS_Statistics_March_2021_-_data_tables.xlsx",
                  "https://www.ons.gov.uk/file?uri=/employmentandlabourmarket/peopleinwork/employmentandemployeetypes/datasets/employmentbyindustryemp13/current/emp13nsafeb2021.xls",
                  "https://www.ons.gov.uk/file?uri=/employmentandlabourmarket/peoplenotinwork/unemployment/datasets/vacanciesbyindustryvacs02/current/vacs02mar2021.xls",
                  "https://www.ons.gov.uk/file?uri=/employmentandlabourmarket/peopleinwork/employmentandemployeetypes/datasets/employmentbyindustryemp13/current/emp13nsafeb2021.xls"]



## 1. Card spending dataset ----------------------------------------------------------


## As first exercise, we can genralise the download url of the first two datasets in order 
## for it to self-adjust for any time a new version of it is released


url_read_cs_1 = requests.get(url_name_list[0])
url_read_cs = url_read_cs_1.text
general_url_name_cs = re.sub('^(.*)(?=ukspendingoncreditanddebitcards/)',"", url_read_cs)
subset_files_cs_1 = re.search('=\"(.*?)xlsx', str(general_url_name_cs)).group(1)
subset_files_cs = subset_files_cs_1 + 'xlsx'
subset_files_clean_cs2 = str("https://www.ons.gov.uk") + str(subset_files_cs) 
resp = requests.get(subset_files_clean_cs2)


## We want to generalise the name of the output too
## for it to self-adjust to the latest updated date of the dataset

output = open(str("raw_data/card_spending_") + str(subset_files_clean_cs2)[-11:], 'wb')
output.write(resp.content)
output.close()




## 2. IoS dataset ----------------------------------------------------------


url_read_i_1 = requests.get(url_name_list[1])
url_read_i = url_read_i_1.text
general_url_name_i = re.sub('^(.*)(?=current/)',"", url_read_i)
subset_files_i_1 = re.search('=\"(.*?)xlsx', str(general_url_name_i)).group(1)
subset_files_i = subset_files_i_1 + 'xlsx'
subset_files_clean_i = str("https://www.ons.gov.uk") + str(subset_files_i) 
resp = requests.get(subset_files_clean_i)
output = open(str("raw_data/ios_data_ios1.xlsx"), 'wb')
output.write(resp.content)
output.close()



## 3. UK Gov Furlough, Employment, and Vacancies dataset ---------------------------



## As additional exercise, we will show how to directly download files from specific links

file_name_list = ["furlough_data", "employment_data", "vacancies_data", "employment_additional"]
  
for i in range(2,6):
    
    if(i == 2): 
        file_extension = ".xlsx"  
    else: 
        file_extension = ".xls"
    
    resp = requests.get(url_name_list[i])
    output = open(str("raw_data/") + str(file_name_list[i-2]) + file_extension, 'wb')
    output.write(resp.content)
    output.close()
```
